```python
rus = 'ёйцукенгшщзхъфывапролджэячсмитьбю'
rus += rus.upper()
eng = '`qwertyuiop[]asdfghjkl;\'zxcvbnm,.~QWERTYUIOP{}ASDFGHJKL:"ZXCVBNM<>'

x = input('Text: ')

er = {e: r for e, r in zip(eng, rus)}
re = {r: e for e, r in zip(eng, rus)}

ans = ''
for i in x:
    if i in er.keys():
        i = er[i]
    elif i in re.keys():
        i = re[i]
    ans += i

print(ans)
```
